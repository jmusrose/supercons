```python
import numpy
import matplotlib.pyplot as plt
```

```python
x = [100, 200, 300, 400, 500, 600]
y1 = [0.918,
0.9277,
0.932,
0.9353,
0.9353,
0.9353,
]
y2 = [0.918,
0.9259,
0.9297,
0.9316,
0.933,
0.9346
]
```

```python
import warnings
warnings.filterwarnings('ignore')
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["axes.unicode_minus"]=False #该语句解决图像中的“-”负号的乱码问题
plt.figure(1, figsize=(12, 6))
plt.plot(x, y1, color = '#FB0000', marker='H', markersize = 10, linewidth = 3)
plt.plot(x, y2, color = '#CA5E00', marker='s', markersize = 10, linewidth = 3)
plt.xlabel('迭代总次数', fontsize = 15)
plt.ylabel('$R^2$', fontsize = 15)
plt.legend(['PSO-XGBoost', 'GA-XGBoost'], loc = 'best', fontsize = 15)
plt.savefig('两个算法速度比较.jpg')
```

```python
import seaborn as sns
```

```python
# 绘制折线图
sns.lineplot(data=y1, color="red")
sns.lineplot(data=y2, color="blue")

# 设置图的标题和坐标轴的标签
plt.title("Two Groups of Data")
plt.xlabel("X")
plt.ylabel("Y")

# 显示图
plt.show()
```
